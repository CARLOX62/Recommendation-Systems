# Text Cleaning/ Text Preprocessing

```python
df['text'] = df['text'] = (
    df['text']
    .str.lower()
    .str.replace(r'[\r\n\t]', ' ', regex=True)   # removes \r, \n, \t
    .str.replace(r'[^\w\s]', '', regex=True)     # removes punctuation
)
```

```python
df["text"][0]
```

```python
df['text'] = df['text'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df["text"][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    stemming = [stemmer.stem(w) for w in tokens]
    return " ".join(stemming)
```

```python
nltk.download('punkt_tab')
tokenization("you are beautiful, beauty")
```

```python
df['text'] = df['text'].apply(lambda x: tokenization(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
```

```python
matrix = tfidvector.fit_transform(df['text'])
```

```python
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
df[df['song'] == 'Green Day']
```

```python
def recommendation(song_df):
    idx = df[df['song'] == song_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for m_id in distances[1:11]:
        songs.append(df.iloc[m_id[0]].song)

    return songs
```

```python
recommendation('In Your Eyes')
```

```python
import pickle
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
```
